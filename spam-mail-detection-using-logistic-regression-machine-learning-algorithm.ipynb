importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

data pre processing

```python
mail_dataset = pd.read_csv('/content/mail_data.csv')
mail_dataset.head()
```

```python
#replace the null values with null string
mail_data  = mail_dataset.where((pd.notnull(mail_dataset)),'')
mail_data.head()
```

replace ham to 1
replace spam to 0

```python
mail_data.loc[mail_data['Category']=='spam', 'Category',] =0
mail_data.loc[mail_data['Category']=='ham', 'Category', ]=1
mail_data.head()
```

now seperating the data as texts and labels

```python
x = mail_data['Message']
y = mail_data['Category']
print(x)
print(y)
```

dividing training data and testing data

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

converting test data into meaningful numerical data (feature extraction)

```python
feature_extraction = TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
x_train_features = feature_extraction.fit_transform(x_train) #we fit here
x_test_features = feature_extraction.transform(x_test)  #we dont fit here

# convert the values as integers in the y values
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

using logistic regression

```python
model = LogisticRegression()
model.fit(x_train_features,y_train)
```

lets evaluate now

```python
prediction_on_training_data = model.predict(x_train_features)
accuracy_on_training_data = accuracy_score(y_train,prediction_on_training_data)
print(accuracy_on_training_data)
```

```python
print('accuracy on training data:')
model.score(x_train_features,y_train) #or u can just use this
```

now evaluating the testing data

```python
prediction_on_testing_data = model.predict(x_test_features)
accuracy_on_testing_data = accuracy_score(y_test,prediction_on_testing_data)
print(accuracy_on_testing_data)
```
